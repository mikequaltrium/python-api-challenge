```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import pandas as pd
import numpy as np
import requests
import time
import csv

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
cities_data = [["City","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp","Wind Speed"]]
for i,city in enumerate(cities):
    url = f"http://api.openweathermap.org/data/2.5/weather?appid={api_key}&q={city}&units=Imperial"
    res = requests.get(url).json()
    length = len(cities)
    if res["cod"] == 200:
        # process the row
        print(f"Processing Record {i} of {length} | {city}")
        cities_data.append([
            res["name"],
            res["clouds"]["all"],
            res["sys"]["country"],
            res["dt"],
            res["main"]["humidity"],
            res["coord"]["lat"],
            res["coord"]["lon"],
            res["main"]["temp_max"],
            res["wind"]["speed"]
        ])
    else:
        # skip the row
        print(f"{city} not found. skipping ...")
```

```python
cities_data

with open("cities_data.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(cities_data)
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
dot_colors = []
for temp in data['Max Temp']:
    if temp > 0:
        color = "blue"
    if temp > 45:
        color = "orange"
    if temp > 80:
        color = "red"
    dot_colors.append(color)
    
legend_elements = [
    Line2D([0], [0], marker='o', color='red', label='Tropic', markerfacecolor='red', markersize=8),
    Line2D([0], [0], marker='o', color='orange', label='Temprate', markerfacecolor='orange', markersize=8),
    Line2D([0], [0], marker='o', color='blue', label='Arctic', markerfacecolor='blue', markersize=8),
]

plt.rcParams["figure.figsize"] = [13,7]
plt.legend(handles=legend_elements, loc='best')
plt.grid()
plt.axvline(x=0,color="g",alpha=0.4)

plt.scatter(data['Lat'],data['Max Temp'],color=dot_colors,alpha=0.5)
plt.title('City Latitude vs. Max Temperature (10/04/2019)')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
# Save the Figure
plt.savefig("city-latitudes-vs-max-temperature.png")
plt.show()
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(data['Lat'],data['Humidity'],color=dot_colors,alpha=0.5)
plt.legend(handles=legend_elements, loc='best')
plt.grid()
plt.title('City Latitude vs. Humidity (10/04/2019)')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.axvline(x=0,color="g",alpha=0.4)
# Save the Figure
plt.savefig("city-latitudes-vs-humidity.png")
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
plt.scatter(data['Lat'],data['Cloudiness'],color=dot_colors,alpha=0.5)
plt.grid()
plt.legend(handles=legend_elements, loc="best")
plt.title('City Latitude vs. Cloudiness (10/04/2019)')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.axvline(x=0,color="g",alpha=0.4)
# Save the Figure
plt.savefig("city-latitudes-vs-cloudiness.png")
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
plt.scatter(data['Lat'],data['Wind Speed'],color=dot_colors,alpha=0.5)
plt.grid()
plt.legend(handles=legend_elements, loc="best")
plt.title('City Latitude vs. Wind Speed (10/04/2019)')
plt.xlabel('Latitude')
plt.ylabel('Wind Speeds (mph)')
plt.axvline(x=0,color="g",alpha=0.4)
# Save the Figure
plt.savefig("city-latitudes-vs-wind-speeds.png")
plt.show()
```

# observable trends based on the data
* The majority of cities with High Tropical temperatures is concentrated heavily between the latitudes of -20 and 20 near the equator. Temperate zones with moderate weather are found further north between the latitudes of 40 and 60 and south between the latitudes of -20 and 50. The highest temperatures logged where mostly between 80 and 100 degrees Fahrenheit and the lowest temperatures where between 0 and 20 degrees Fahrenheit. 

* The warmer tropical zones tend to have a dryer climate with less than 80% Humidity

* Warmer zones near the equator also tend to have less wind (less than 20 mph) while colder climates have greater wind speeds (up to 37mph) 

* Cloudiness does not demonstrate any apparent trends based on latitude.
